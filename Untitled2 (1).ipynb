Подключим все необходимые библиотеки

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
import statistics as st
import numpy as np
alpha = 0.05
```

```python
df = pd.read_excel("FPS_hw_x_df_x_taxi.xlsx")
```

```python
df
```

Сосздадим две отдельные таблицы (2 выборки)

```python
df_C = df[df["offer_class_group"] == "Comfort"]
df_E = df[df["offer_class_group"] == "Economy"]
print('Длинна выборок равна', len(df_C) == len(df_E))
```

Построение графика плотности распределения дистанции поездок

```python
sns.histplot(df_C["distance_km"], kde = True)
sns.histplot(df_E["distance_km"], kde = True)
plt.title("плотности распределения дистанций поездок")
plt.legend(["Comfort", "Economy"])
plt.show()
```

Исходя из построенных гистограмм, можем утверждать, что оба распределения являются логнормальными

H0: Нет разницы между предпочтениями в классах в зависимости от дальности поездки

H1: На дальние поездки пользователи предпочитают Economy-классу более комфортные условия поездки в Comfort-классе

Проверка для не нормально-распределённых выборок

```python
result = mannwhitneyu(df_C["distance_km"], df_E["distance_km"])
print("p-value: %.8f" % result.pvalue)
if (result.pvalue < alpha):
  print("Отклоняем H0")
else:
  print("Не отклоняем H0")
```

```python
#Предпочтения заказа Comfort-класса статистически значимо отличаются от Economy-класса
```

Проверка для нормально-распределённых выборок

```python
result = ttest_ind(df_C["distance_km"], df_E["distance_km"])
print("p-value: %.8f" % result.pvalue)
if (result.pvalue < alpha):
  print("Отклоняем H0")
else:
  print("Не отклоняем H0")
```

```python
#Предпочтения заказа Comfort-класса статистически значимо отличаются от Economy-класса
```
